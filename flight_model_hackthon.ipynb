#### Use Randomforest

```python
forest = RandomForestRegressor()
```

```python
new_df = df[["Revenue (USD)","Operating Cost (USD)","Revenue per ASK","Cost per ASK","Profit (USD)"]]
new_df
```

```python
x_train,x_test,y_train,y_test=train_test_split(new_df.drop(columns=["Profit (USD)"]),new_df["Profit (USD)"],test_size=0.3,random_state=42)
```

```python
x_train
```

```python
forest=RandomForestRegressor()
```

```python
forest.fit(x_train,y_train)
```

```python
y_pred = forest.predict(x_test)
```

```python
print(r2_score(y_test,y_pred)*100)
```

```python
print(mean_squared_error(y_test,y_pred)*100)
```

```python
print(mean_absolute_error(y_test,y_pred)*100)
```

```python
y_pred2 = forest.predict(x_train)
```

```python
print(r2_score(y_pred2,y_train)*100)
```

```python
scores = cross_val_score(forest, x_train, y_train, cv=5, scoring='r2')
print("r2_score:", scores)
```
